```python
import codecs

dataset = list()
file_avg = codecs.open(r'avg_weather.txt', 'r', 'utf-8')
for line in file_avg:
    lst = line.strip().split('\t')
    lst[2] = float(lst[2])
    lst[3] = float(lst[3])
    dataset.append(lst)
dataset[0], len(dataset)
```

```python
from pyecharts import Bar, Line
```

```python
# 平均昼夜温差排行
dataset_avg_wencha = sorted(dataset, key=lambda x:abs(x[2]-x[3]))
dataset_avg_wencha[:3]
```

```python

bar_avg_desc = Bar("平均昼夜温差排行")
bar_avg_desc.add("平均昼夜温差-升序", [x[1] for x in dataset_avg_wencha[:20]], 
        [x[2]-x[3] for x in dataset_avg_wencha[:20]], xaxis_interval=0, xaxis_rotate=45)

bar_avg_desc
# 昼夜温差最小前20的大多数都是南方城市，有少部分是其他地区的沿海城市
```

```python
bar_avg_asc = Bar("平均昼夜温差排行")
bar_avg_asc.add("平均昼夜温差-降序", [x[1] for x in dataset_avg_wencha[-1:-20:-1]], 
        [x[2]-x[3] for x in dataset_avg_wencha[-1:-20:-1]], xaxis_interval=0, xaxis_rotate=45)

bar_avg_asc
# 昼夜温差最大的前20大多数都是北方城市，有少数是山区和盆地
```

```python
dataset_avg_max = sorted(dataset, key=lambda x:abs(x[2]))
dataset_avg_max[:3]
```

```python
bar_avgmax_asc = Bar("平均最高气温排行")
bar_avgmax_asc.add("平均最高气温-降序", [x[1] for x in dataset_avg_max[-1:-20:-1]], 
        [x[2] for x in dataset_avg_max[-1:-20:-1]], xaxis_interval=0, xaxis_rotate=45)
bar_avgmax_asc.add("平均最低气温", [x[1] for x in dataset_avg_max[-1:-20:-1]], 
        [x[3] for x in dataset_avg_max[-1:-20:-1]], xaxis_interval=0, xaxis_rotate=45)

bar_avgmax_asc
# 平均最高气温较高的城市集中在南方沿海城市和山区，这些城市的平均最低气温也较高，属于炎热城市
```

```python
bar_avgmax_asc = Bar("平均最高气温排行")
bar_avgmax_asc.add("平均最高气温-升序", [x[1] for x in dataset_avg_max[:20]], 
        [x[2] for x in dataset_avg_max[:20]], xaxis_interval=0, xaxis_rotate=45)
bar_avgmax_asc.add("平均最低气温", [x[1] for x in dataset_avg_max[:20]], 
        [x[3] for x in dataset_avg_max[:20]], xaxis_interval=0, xaxis_rotate=45)

bar_avgmax_asc
# 平均最高气温较低的城市集中在北方和高原地区，这些城市的平均最低气温也较低，属于寒冷城市
```

```python
# 平均气温排行
dataset_avg = sorted(dataset, key=lambda x:(x[2]+x[3])/2)
dataset_avg[:3]
```

```python
bar_avg_asc = Bar("平均气温排行")
bar_avg_asc.add("平均气温-降序", [x[1] for x in dataset_avg[-1:-20:-1]], 
        [(x[2]+x[3])/2 for x in dataset_avg[-1:-20:-1]], xaxis_interval=0, xaxis_rotate=45)
bar_avg_asc
```

```python
bar_avg_asc = Bar("平均气温排行")
bar_avg_asc.add("平均气温-升序", [x[1] for x in dataset_avg[:20]], 
        [(x[2]+x[3])/2 for x in dataset_avg[:20]], xaxis_interval=0, xaxis_rotate=45)
bar_avg_asc
```

```python
dataset_type = list()
file_type = codecs.open(r'weathertype.txt', 'r', 'utf-8')
for line in file_type:
    lst = line.strip().split('\t')
    lst[2] = float(lst[2])
    lst[3] = float(lst[3])
    dataset_type.append(lst)
dataset_type[0], len(dataset_type)
```

```python
# 晴天数
dataset_type_sun = sorted(dataset_type, key=lambda x:x[2])
dataset_type_sun[:3]

bar_sun_asc = Bar("晴天数排行")
bar_sun_asc.add("晴天数-降序", [x[1] for x in dataset_type_sun[-1:-20:-1]], 
        [x[2] for x in dataset_type_sun[-1:-20:-1]], xaxis_interval=0, xaxis_rotate=45)
bar_sun_asc
```

```python
# 雨天数
dataset_type_rain = sorted(dataset_type, key=lambda x:x[3])
dataset_type_rain[:3]

bar_rain_asc = Bar("雨天数排行")
bar_rain_asc.add("雨天数-降序", [x[1] for x in dataset_type_rain[-1:-20:-1]], 
        [x[3] for x in dataset_type_rain[-1:-20:-1]], xaxis_interval=0, xaxis_rotate=45)
bar_rain_asc
```
